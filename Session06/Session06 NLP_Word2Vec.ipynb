```python
import nltk
#nltk.download('movie_reviews')
#nltk.download('punkt')
```

```python
from nltk.corpus import movie_reviews
sentences = [list(s) for s in movie_reviews.sents()]
```

```python
# Word2Vec 모델 만들기
from gensim.models.word2vec import Word2Vec
model = Word2Vec(sentences,size=100, window=3, min_count=2, sg=1)
model.init_sims(replace=True)
```

```python
print(model.similarity('actor', 'actress'))
print(model.similarity('he', 'she'))
print(model.similarity('actor', 'she'))
```

```python
model.most_similar("hero")
```

```python
model.most_similar(positive=['actor', 'he'], negative='actress', topn=1)
```

#### Korean

```python
import codecs
from bs4 import BeautifulSoup
from konlpy.tag import Okt
from gensim.models import word2vec
```

```python
# utf-16 인코딩으로 파일을 열고 글자를 출력하기 
fp = codecs.open("C:\\Users\\admin\\Desktop\\BRJO0443.txt", "r", encoding="utf-16")
soup = BeautifulSoup(fp, "html.parser")
body = soup.select_one("text > body")
text = body.getText()
```

```python
# 텍스트를 한 줄씩 처리하기 
twitter = Okt()
results = []
lines = text.split("\r\n")
for line in lines:
    # 형태소 분석하기 
    # 단어의 기본형 사용
    malist = twitter.pos(line, norm=True, stem=True)
    r = []
    for word in malist:
        # 어미/조사/구두점 등은 대상에서 제외 
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0])
    rl = (" ".join(r)).strip()
    results.append(rl)
    print(rl)
```

```python
# 파일로 출력하기  
txt_file = 'europe.txt'
with open(txt_file, 'w', encoding='utf-8') as fp:
    fp.write("\n".join(results))
```

```python
# Word2Vec 모델 만들기 
data = word2vec.LineSentence(txt_file)
model = word2vec.Word2Vec(data, 
    size=200, window=10, hs=1, min_count=2, sg=1)
model.save("europe.model")
```

```python
model = word2vec.Word2Vec.load("europe.model")
model.most_similar("파리")
```
